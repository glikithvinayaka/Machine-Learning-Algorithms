```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df = pd.read_csv('shopping_data.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
data = df.iloc[:, 2:5]
data.head()
```

```python
#x = shc.linkage(data)
```

```python
plt.figure(figsize=(15,7))
shc.dendrogram(shc.linkage(data, method='ward'));
```

```python
clusters = AgglomerativeClustering(n_clusters= 3, affinity='euclidean', linkage='ward')
clusters.fit_predict(data)
```

```python
clusters.labels_
```

```python
data['Segment'] = clusters.labels_
```

```python
data.sample(10)
```

```python
data.Segment.value_counts()
```

```python
data.isnull().sum()
```

```python
plt.scatter(data['Spending Score (1-100)'], data['Annual Income (k$)'], c = clusters.labels_)
plt.xlabel('Spending Score')
plt.ylabel('Annual Income')
```
